Y por consiguiente, si existen diferencias se deberá a que el dato es por inverter.

```python
temp = p1g[p1g.SOURCE_KEY.isin(seleccion)].set_index('DATE_TIME')
temp
```

En los datos ya se aprecian las diferencias, pero se comprobará sobre más datos para evitar posibles efectos de esos registros en concreto.

Se comprobará gráficamente, y por simplificar se cogerá solo una muestra de días.

```python
temp = temp.loc['2020-06-01':'2020-06-05']
temp
```

```python
temp.reset_index(inplace=True)
plt.figure(figsize = (16,8))
sns.lineplot(data = temp, x = 'DATE_TIME' , y = 'DAILY_YIELD', hue = 'SOURCE_KEY');
```

Definitivamente diferentes inverters tienen diferentes datos en el mismo momento temporal, por lo que esta variable es **por inverter**.

## CREACIÓN DEL DATAMART ANALITICO

### Unión de los datasets de generación

```python
gener = pd.concat([p1g,p2g],axis = 'index')
```

Renombrado de las variables para hacerlas más descriptivas y usables.

```python
gener.columns = ['fecha','planta','inverter_id','kw_dc','kw_ac','kw_dia','kw_total']
gener
```

Análisis de coherencia para comprobar, como indica la documentación recibida, que los datos de `kw_dia` y `kw_total` están directamente relacionados con `kw_dc` y `kw_ac`.

```python
gener2 = gener.copy()
```

Creación de variable `date` para agregar por ella.

```python
gener2['date'] = gener2.fecha.dt.date
```

La suma por planta, date e inverter de kw_dc o de kw_ac debería coincidir con el máximo de kw_dia.

```python
gener2 = gener2.groupby(['planta','date','inverter_id']).agg({'kw_dc':sum,
                                                              'kw_ac':sum,
                                                              'kw_dia':max,
                                                              'kw_total':max}).reset_index()
gener2
```

Ordenación para el análisis.

```python
gener2 = gener2.sort_values(['planta','inverter_id','date'])
gener2
```

`Kw_dia` no concuerda ni con `kw_dc` ni con `kw_ac`.

Comprobación para ver si concuerda con `kw_total`, para ello se calculará el incremento diario de `kw_total` que debería coincidir con el máximo de `kw_dia` del día anterior.

```python
gener2['lag1'] = gener2.groupby(['planta','inverter_id']).kw_total.shift(1)
gener2['incremento'] = gener2.kw_total - gener2.lag1
```

Comprobamos en la planta 1.

```python
gener2[gener2.planta == 'p1'].head(50)
```

Comprobamos en la planta 2.

```python
gener2[gener2.planta == 'p2'].head(50)
```

Conclusiones:
* `kw_dia` tiene coherencia con `kw_total`.
* Pero éstas no tienen coherencia con `kw_dc` ni con `kw_ac`.
* Parece que estuvieran en diferentes unidades o hubiera algún cálculo intermedio que se desconoce.
* por tanto se diferenciarán 2 bloques a poder usar: o bien `kw_dc` con `kw_ac`, o bien `kw_dia` con `kw_total`, pero no se mezclarán entre sí.

### Unión de los datasets de mediciones ambientales

```python
temper = pd.concat([p1w,p2w], axis = 'index')
temper
```

Renombrado de las variables para hacerlas más descriptivas y usables.

```python
temper.columns = ['fecha','planta','sensor_id','t_ambiente','t_modulo','irradiacion']
```

### Creación del datamart analitico

```python
df = pd.merge(left = gener, right = temper, how = 'left', on = ['fecha','planta'])
```

Comprobación de nulos:

```python
df.isna().sum()
```

Identificación de posibles patrones en los valores nulos.

```python
nulos = df[df.sensor_id.isna()]
nulos
```

Se trata del día 3 de Junio a las 14:00, que por algún motivo no tiene datos de temperatura pero solo para 4 inverters de la planta 1.

Se buscará en el dataset de temperatura si existe ese datetime.

```python
temper[temper.fecha.between('2020-06-03 13:30:00', '2020-06-03 14:30:00')]
```

Efectivamente vemos que falta ese tramo en ambas plantas. Pero sin embargo solo hay mediciones en esa hora en la planta 1, y solo en 4 inverters.

Por tanto habría dos posibles soluciones:
* Imputar esos datos para esos invertes.
* Eliminar esos 4 registros.

Dado que parece una franja de medición propia solo de 4 inverters de la planta 1 se opta por eliminarlos.

```python
df.dropna(inplace = True)
```

Fecha como index.

```python
df.set_index('fecha', inplace = True)
```
